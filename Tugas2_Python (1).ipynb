```python
import numpy as np

# jumlahkan bilangan 1 sampai N
N = 100

xs = np.arange(1, N+1)
total = np.sum(xs)

print("Numbers:", xs)
print("Sum:", total)
```

```python
%%writefile compl.in
x y
0.6 -6.2
1.6 9.3
1.8 16.
import numpy as np

# baca file (skip header)
A = np.loadtxt("compl.in", skiprows=1)

print("Full Array:")
print(A)

# pisahkan kolom
xs = A[:, 0]
ys = A[:, 1]

print("xs:", xs)
print("ys:", ys)

# simpan ulang ke file
np.savetxt("compl4.out", A)

print("File compl4.out berhasil dibuat.")
```
