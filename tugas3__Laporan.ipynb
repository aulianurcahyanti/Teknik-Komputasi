##  Masalah Fisika

Dalam suatu sistem fisika, terdapat empat gaya (F1, F2, F3, F4)
yang bekerja pada sebuah benda dalam kondisi setimbang.

Berdasarkan hukum keseimbangan gaya (ΣF = 0), hubungan antar gaya
dapat dinyatakan dalam bentuk sistem persamaan linear.

Untuk menentukan besar masing-masing gaya, digunakan metode
Gauss-Jordan dengan pivoting.

```python
import numpy as np

def gauss_jordan(A, b):
    n = len(b)
    M = np.hstack((A.astype(float), b.reshape(-1,1)))

    for k in range(n):
        # Pivoting
        max_row = np.argmax(abs(M[k:, k])) + k
        M[[k, max_row]] = M[[max_row, k]]

        # Normalisasi
        M[k] = M[k] / M[k, k]

        # Eliminasi
        for i in range(n):
            if i != k:
                M[i] = M[i] - M[i, k] * M[k]

    return M[:, -1]

A = np.array([
    [1, 1, 1, 1],
    [2, 1, 1, 1],
    [1, 2, 1, 1],
    [1, 1, 2, 1]
])

b = np.array([12, 15, 16, 17])

solusi = gauss_jordan(A, b)

print("Solusi:")
for i in range(len(solusi)):
    print(f"F{i+1} =", round(solusi[i], 2))
```
