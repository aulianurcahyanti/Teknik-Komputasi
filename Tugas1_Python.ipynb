```python
sentence = "this is a rambling sentence that simply goes on and on and on and just simply will not stop that is just the way things are"

words = sentence.split()

wordtocount = {}

# hitung populasi kata
for word in words:
    wordtocount[word] = wordtocount.get(word, 0) + 1

# tampilkan tabel
for word in wordtocount:
    print(word, ":", wordtocount[word])

# hitung berapa kata muncul 1x, 2x, 3x
once = twice = thrice = 0

for count in wordtocount.values():
    if count == 1:
        once += 1
    elif count == 2:
        twice += 1
    elif count == 3:
        thrice += 1

print("Words appearing once:", once)
print("Words appearing twice:", twice)
print("Words appearing thrice:", thrice)

# cek total
total_words = len(words)
check_total = once + 2*twice + 3*thrice
print("Check total:", total_words == check_total)
```

```python
def sum_odds(n):
    total = 0
    for i in range(1, n+1, 2):
        total += i
    return total

while True:
    n = int(input("Enter a large odd number (even to stop): "))

    if n % 2 == 0:
        print("Program stopped.")
        break

    print("Sum:", sum_odds(n))
```

```python
%%writefile odd_in.dat
9
15
21
7
def sum_odds(n):
    total = 0
    for i in range(1, n+1, 2):
        total += i
    return total

with open("odd_in.dat", "r") as fin, open("odd_out.dat", "w") as fout:
    counter = 1
    for line in fin:
        n = int(line.strip())
        total = sum_odds(n)
        fout.write(f"{counter} {n} {total}\n")
        counter += 1
```

```python
def read_xy(filename):
    xs = []
    ys = []

    with open(filename, "r") as f:
        for line in f:
            parts = line.split()
            x = float(parts[0])
            y = float(parts[1])
            xs.append(x)
            ys.append(y)

    return xs, ys


# baca file
xvals, yvals = read_xy("xy.dat")

# hitung jumlah & rata-rata y
total_y = sum(yvals)
avg_y = total_y / len(yvals)

print("Total y:", total_y)
print("Average y:", avg_y)
```
