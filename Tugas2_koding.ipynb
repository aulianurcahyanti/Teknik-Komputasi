```python
def f(x):
  return x**2 - x - 2
```

Bisection

```python
def bisection(f, a, b, tol=1e-6, max_iter=100):
    if f(a)*f(b) > 0:
        print("Interval tidak valid")
        return None

    for i in range(max_iter):
        c = (a + b) / 2

        if abs(f(c)) < tol:
            return c

        if f(a)*f(c) < 0:
            b = c
        else:
            a = c

    return c
```

```python
print(bisection(f, 1, 3))
```

regulafalsi

```python
def regulafalsi(f, a, b, tol=1e-6, max_iter=100):
    for i in range(max_iter):
        c = (a*f(b) - b*f(a)) / (f(b) - f(a))

        if abs(f(c)) < tol:
            return c

        if f(a)*f(c) < 0:
            b = c
        else:
            a = c

    return c
```

```python
print(regulafalsi(f, 1, 3))
```

secant

```python
def secant(f, x0, x1, tol=1e-6, max_iter=100):
    for i in range(max_iter):
        x2 = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))

        if abs(f(x2)) < tol:
            return x2

        x0 = x1
        x1 = x2

    return x2
```

```python
print(secant(f, 1, 3))
```

newtonrap

```python
def df(x):
    return 2*x - 1

def newtonrap(f, df, x0, tol=1e-6, max_iter=100):
    for i in range(max_iter):
        x1 = x0 - f(x0)/df(x0)

        if abs(f(x1)) < tol:
            return x1

        x0 = x1

    return x1
```

```python
print(newtonrap(f, df, 2))
```

```python
def regulafalsi_table(f, a, b, tol=1e-6, max_iter=20):
    print("Iterasi | a | b | c | f(c)")

    for i in range(max_iter):
        c = (a*f(b) - b*f(a)) / (f(b) - f(a))

        print(i+1, "|", round(a,6), "|", round(b,6), "|", round(c,6), "|", round(f(c),6))

        if abs(f(c)) < tol:
            break

        if f(a)*f(c) < 0:
            b = c
        else:
            a = c
```

```python
regulafalsi_table(f, 1, 3)
```

```python
def secant_table(f, x0, x1, tol=1e-6, max_iter=20):
    print("Iterasi | x0 | x1 | x2 | f(x2)")

    for i in range(max_iter):
        x2 = x1 - f(x1)*(x1 - x0)/(f(x1) - f(x0))

        print(i+1, "|", round(x0,6), "|", round(x1,6), "|", round(x2,6), "|", round(f(x2),6))

        if abs(f(x2)) < tol:
            break

        x0 = x1
        x1 = x2
```

```python
secant_table(f, 1, 3)
```

```python
def g(x):
    return (x + 2)**0.5
```

```python
def fixed_point(g, x0, tol=1e-6, max_iter=20):
    print("Iterasi | x")

    for i in range(max_iter):
        x1 = g(x0)
        print(i+1, "|", round(x1,6))

        if abs(x1 - x0) < tol:
            break

        x0 = x1

    return x1
```

```python
print(fixed_point(g, 1))
```
